## **Habilitando paquetes**

```r
library(data.table)
```

## **Creando una función para leer en Google Drive, de carpeta pública**

```r
read.csv.googledrive <- function(url, sep=",", dec=".") {
  require(data.table)
  id <- strsplit(url,"d/")[[1]][2]
  id <- strsplit(id,"/view")[[1]][1]
  return(fread(sprintf("https://docs.google.com/uc?id=%s&export=download", id), sep=sep, dec=dec, integer64 = "character"))}
```

## **Obteniendo datos**

```r
vinculo = "https://drive.google.com/file/d/1Iyr5hD9DbYQ_His3HFJ0fbAkhdOyzaoK/view?usp=sharing"
```
