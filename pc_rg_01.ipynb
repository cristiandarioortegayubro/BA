## **Conjunto de Datos**

```
url = "https://raw.githubusercontent.com/cristiandarioortegayubro/BA/main/Datasets/publicidad_multiple.csv"
```

## ***Dataframe***

```
datos = pd.read_csv(url, index_col = 0)
```

```
datos.head(10)
```

## **Configuración del entorno PyCaret**

```
regresion = setup(data = datos, 
                  target = "ventas",
                  train_size = 0.80, 
                  session_id=123)
```
