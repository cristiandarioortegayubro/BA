## **Generación del Panel de Datos**

***Con la creación de una lista, se seleccionan las acciones que forman parte del portfolio, estas son:***

+ ***Banco Macro***
+ ***Alua***
+ ***Edenor***
+ ***YPF***
+ ***Pampa***

```python
acciones = ["BMA.BA","ALUA.BA","EDN.BA", "YPFD.BA", "PAMP.BA"]
```

***Una vez definida la lista, se obtienen los datos remotos de cada acción, a traves de Data Reader. Lo que se va a obtener por resultado, es una estructura de datos multidimensional, es decir, un panel de datos de tres dimensiones***

```python
portfolio = round(dtr.DataReader(acciones, data_source="yahoo", start="2021-12-30"),1)
portfolio
```

```python
portfolio.axes
```

### **Creando DataFrame del Panel de Datos**

#### **Banco Macro**

```python
BMA = portfolio.xs("BMA.BA", level="Symbols", axis=1)
BMA
```

#### **Alua**

```python
ALUA = portfolio.xs("ALUA.BA", level="Symbols", axis=1)
ALUA
```

#### **Edenor**

```python
EDENOR = portfolio.xs("EDN.BA", level="Symbols", axis=1)
EDENOR
```

### **Obteniendo una serie del Panel de Datos**

```python
precios = round(portfolio["Close"],2)
precios.rename(columns={"BMA.BA":"Banco Macro",
                        "ALUA.BA":"Aluar",
                        "EDN.BA":"Edenor",
                        "YPFD.BA":"YPF", 
                        "PAMP.BA":"Pampa"},inplace=True)
precios
```

#### **Calculando el cambio porcentual diario de las cotizaciones**

```python
rendimientos = round(precios.pct_change(),4)
rendimientos.dropna(how="any", inplace=True)
rendimientos.reset_index(inplace=True)
rendimientos
```

```python
rendimientos.info()
```

#### **Calculando las diferencias de precio diarias de las cotizaciones**

```python
diferencia = round(precios.diff(),4)
diferencia.dropna(how="any", inplace=True)
diferencia.reset_index(inplace=True)
diferencia
```

#### **Análisis estadístico descriptivo**

```python
round(rendimientos.describe(),5)
```

```python
round(rendimientos.describe(),5).T
```

## **Optimización de cartera**

### **Rendimientos esperados**

Los modelos de optimización requieren conocer previamente los rendimientos esperados. En la práctica, estos se calculan en función de datos históricos. 

El módulo `expected_returns` de la biblioteca `PyPortfolioOpt` proporciona las funciones para estimar los rendimientos esperados de las acciones.

```python
rendEsp = expected_returns.mean_historical_return(precios, returns_data=False, compounding=False, log_returns=False)
rendEsp
```

### **Riesgos**

El módulo `risk_models` de la `bibliotecaPyPortfolioOpt` proporciona las funciones para estimar el riesgo de las acciones. El método más usado es el que estima la matriz de varianzas y covarianzas en función de los datos históricos. Esta matriz es una representación del riesgo de las acciones individuales (diagonal) y de la covariación entre acciones.

```python
matrizCov = risk_models.sample_cov(precios)
matrizCov
```

```python
plt.figure(figsize=(10,9))
sns.heatmap(rendimientos.corr(), cmap="RdBu", linewidth=0.5, annot=True, vmax=0.8)
plt.show()
```

### **Armar la frotnera eficiente**

:La función `EfficientFrontier` permite definir la frontera eficiente, es decir, el conjunto de carteras que maximiza el rendimiento para determinados niveles de riesgo o minimiza el riesgo para determinados niveles de rendimiento.

El objeto creado con la función contiene varios métodos que corresponden a las diferentes funciones objetivo a optimizar:

* `efficient_risk()` : maximiza el rendimiento dado un nivel de riesgo.
* `efficient_return()` : minimiza el riesgo dado un nivel de rendimiento.
* `max_sharpe()` : optimiza para obtener el máximo ratio de Sharpe, también conocida como cartera de tangencia.

**Máximo rendimiento dado un nivel de riesgo**

Determina las participaciones que maximizan el rendimiento para un nivel dado de riesgo. Es uno de los supuestos básicos de la teoría de Markowitz (Insaciabilidad). 

```python
fe = EfficientFrontier(rendEsp, matrizCov)
fe.efficient_risk(target_volatility = 0.35)
pesos = fe.clean_weights() 
pesos
```

```python
datos = pd.DataFrame(pesos, index = [0])
datos = datos.T
datos.columns = ["Participacion"]
datos
```

```python
px.pie(datos, 
       values="Participacion", 
       names=datos.index, 
       title="Máximo rendimiento dado un riesgo del 35%",
       hole=.4,
       color_discrete_sequence = px.colors.sequential.Turbo)
```

```python
fe.portfolio_performance(verbose=True)
```

**Mínimo riesgo dado un nivel de rendimiento**

Determina las participaciones que minimizan el riesgo para un nivel dado de rendimiento. Es uno de los supuestos básicos de la teoría de Markowitz (Aversión al riesgo). 

```python
fe = EfficientFrontier(rendEsp, matrizCov)
fe.efficient_return(target_return= 0.4)
pesos = fe.clean_weights() 
pesos
```

```python
datos = pd.DataFrame(pesos, index = [0])
datos = datos.T
datos.columns = ["Participacion"]
datos
```

```python
px.pie(datos, 
       values="Participacion", 
       names=datos.index, 
       title="Minimo riesgo dado un nivel de rendimiento del 40%",
       hole=.4,
       color_discrete_sequence = px.colors.sequential.Bluered)
```

```python
fe.portfolio_performance(verbose=True)
```

**Máximo ratio de Sharpe**

Devuelve los pesos de las acciones que maximizan el ratio de Sharpe. Este método intenta determinar los pesos o las participaciones de las acciones que maximize el rendimiento por unidad de riesgo. Esta cartera es conocida como la cartera de tangencia, es la cartera donde la linea de mercado de capitales (capital market line) es tangente a la frontera eficiente.

```python
fe = EfficientFrontier(rendEsp, matrizCov)
pesos = fe.max_sharpe()
pesos = fe.clean_weights()
pesos
```

```python
datos = pd.DataFrame(pesos, index = [0])
datos = datos.T
datos.columns = ["Participacion"]
datos
```

```python
px.pie(datos, 
       values="Participacion", 
       names=datos.index, 
       title="Máximo ratio de Sharpe",
       hole=.4,
       color_discrete_sequence = px.colors.sequential.Cividis)
```

```python
fe.portfolio_performance(verbose=True)
```
