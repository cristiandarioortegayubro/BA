# **Detección de Anomalías - SVM**


## **Carga de librerías necesarias**

### **Analisis de datos**

```
import numpy as np
import pandas as pd
```

### **Modelo y preprocesamiento**

```
from sklearn.svm import OneClassSVM
```

### **Graficos**

```
import plotly.express as px
```

## **Obtención de los Datos**

### **Generación de Datos**

```
d1 = np.random.multivariate_normal(mean = np.array([-.5, 0]), cov = np.array([[1, 0], [0, 1]]), size = 100)
```

```
d2 = np.random.multivariate_normal(mean = np.array([15, 10]), cov = np.array([[1, 0.3], [.3, 1]]), size = 100)
```

```
outliers = np.array([[0, 10],[0, 9.5]])
```

### **Creación del dataframe**

```
df = pd.DataFrame(np.concatenate([d1, d2, outliers], axis = 0), columns = ['Var 1', 'Var 2'])
```

```
df
```

## **Modelo**

### **Estableciendo el algoritmo**

```
svm = OneClassSVM(kernel = 'rbf', 
                  degree = 5, 
                  coef0 = 0.0,
                  gamma = 0.1, 
                  tol = 0.001, 
                  nu = 0.01, 
                  shrinking = True, 
                  cache_size = 200, 
                  verbose = False, 
                  max_iter = - 1)
```

### **Aplicando el algoritmo al dataframe**

```
svm.fit(df)
```

### **Realizando la predicción**

```
prediccion = svm.predict(df)
```

```
prediccion
```

### **Puntuacion**

```
score = svm.score_samples(df)
```

## **Graficos del modelo**

### **Gamma 0.1**

```
px.scatter(df,
           x='Var 1', 
           y='Var 2',
           color=score,
           color_continuous_scale='Inferno',
           template="simple_white",
           labels={"Var 1":"Variable 1",
                   "Var 2":"Variable 2",
                   "color":"Score"})
```

```
px.scatter(df,
           x='Var 1', 
           y='Var 2',
           color=prediccion,
           color_continuous_scale='Bluered_r',
           template="simple_white",
           labels={"Var 1":"Variable 1",
                   "Var 2":"Variable 2",
                   "color":"Score"})
```

### **Gamma 5**

```
svm = OneClassSVM(kernel = 'rbf', 
                  degree = 5, 
                  coef0 = 0.0,
                  gamma = 5, 
                  tol = 0.001, 
                  nu = 0.01, 
                  shrinking = True, 
                  cache_size = 200, 
                  verbose = False, 
                  max_iter = - 1)
```

```
svm.fit(df)
prediccion = svm.predict(df)
score = svm.score_samples(df)
```

```
px.scatter(df,
           x='Var 1', 
           y='Var 2',
           color=score,
           color_continuous_scale='Inferno',
           template="simple_white",
           labels={"Var 1":"Variable 1",
                   "Var 2":"Variable 2",
                   "color":"Score"})
```

```
px.scatter(df,
           x='Var 1', 
           y='Var 2',
           color=prediccion,
           color_continuous_scale='Bluered_r',
           template="simple_white",
           labels={"Var 1":"Variable 1",
                   "Var 2":"Variable 2",
                   "color":"Score"})
```
